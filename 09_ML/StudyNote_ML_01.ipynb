### Decision Tree_2
- Decision Tree 분할기준
    - Split Criterion
- Information Gain, 정보 획득
    - 정보의 가치를 반환하는데 발생하는 사전의 확률이 작을수록 정보의 가치는 커진다.
    - 어떤 속성을 선택함으로 인해 데이터를 더 잘 구분하게 되는 것
- Entropy
    - 열역학 용어로 물질의 열적 상태를 나타내는 물리량 단위 중 하나
    - 무질서 정도
    - disorder
    - uncertainty
    - 확률 분포 무질서도나 불확실성 혹은 정보 부담 정도를 나타내는 정보 엔트로피 개념을 고안함_1948년 클로드 섀년

```python
import numpy as np

p = np.arange(0.001, 1, 0.001)
plt.grid()
plt.title('$-p \log_{2}{p}$')
plt.plot(p, -p * np.log2(p ))
plt.show()
```

### Decision Tree_3
- $-p_ilog_2p_i$
    - p는 해당 데이터 해당 클래스에 속할 확률
    - 어떤 확률 분포로 일어나는 사건을 표현하는데 필요한 정보의 양, 이것이 커질수록 확률 분포 불확실성이 커지며 결과에 대한 예측이 어려워짐
- $Entropy = \sum^m_{k=1}-p_ilog_2p_i = \sum^d_{i=1}R_i \sum^m_{k=1}-p_ilog_2p_i$
    - 분할하여 각 엔트로피를 구한 후, 분할 확률을 곱해 더한 것이 분할전보다 작다면 엔트로피가 감소했으므로 분할하는 것이 좋다
- $Gini = \sum^d_{k=1}R_i\{1-\sum^m_{k=1}p^2_{ik}\}$
    - 지니 계수
    - Gini Index
    - 불순도율
    - 엔트로피 계산량이 많아, 비슷한 개념으로 계산량이 적어 활용도가 높다
- 컴퓨터가 잘게 선을 나누어 각 엔트로피를 비교함

### Scikit Learn_DecisionTreeClassifier
1. `from sklearn.tree import DecisionTreeClassifier`
	- scikit learn에서 tree계열 중 특정 알고리즘을 가져옴
2. `iris_tree = DecisionTreeClassifier()`
    - data명_clf
        - clf는 classfier로 분류라는 의미
    - () 안에 수많은 옵션 지정이 있음, 현재는 생략
3. `iris_tree.fit(iris.data[:, 2:], iris.target)`
    - fit 명령
        - 데이터를 줄테니 학습하시오
        - 첫 번째 인수
            - 데이터
        - 두 번째 인수
            - 정답
- `from sklearn.metrics import accuracy_score`
	- accuracy를 계산해주는 함수
	- `y_pred_tr = iris_tree.predict(iris.data[:, 2:])`
		- predict 명령
			- 데이터를 줄테니 예측해봐라
	- `accuracy_score(iris.target, y_pred_tr)`
		- 첫 번째 인수
			- 정답 데이터
		- 두 번재 인수
			- 예측 데이터
		- 결과값
			- 정확도

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
print(accuracy_score(iris.target, y_pred_tr))
```

### 데이터 나누기
- `from sklearn.model_selection import train_test_split`
    - train, test 데이터로 나누는 모듈
- `train_test_split(활용 데이터, 정답 데이터, test_size = , stratify = , random_state = )`
    - test_size
        - 시험용 데이터를 전체에 0.2만큼 사용하는 것
    - 무작위성
        - random_state로 고정
        - 학습 일관성이 목적
    - stratify
        - 분류 비율을 맞추는 옵션, 데이터 분리할 때 추가
    - 옵션여부, 가중치값, 부여하는 순간 등은 모두 디자이너에 달려있음
- `np.unique(궁금한 데이터, return_counts = True)`
    - 각 그룹 분포가 어떻게 되어있는지 확인하는 것
    - return_counts 옵션을 활성화해야 각 그룹마다 몇 개가 있는지 반환해줌
    - y_test 등 y 값을 확인함


```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

# 활용 데이터
# 평면에 데이터 그리기 위해 일단 2개만 활용
features = iris.data[:, 2:]
# 정답 data
labels = iris.target

# test_size는 시험용 데이터를 전체에 0.2만큼 사용하는 것
# random_state로 고정
X_train, X_test, y_train, y_test = train_test_split(features,
                                                   labels,
                                                   test_size = 0.2,
                                                   random_state =13
                                                   )
```

```python
print(X_train.shape, X_test.shape)
```

```python
import numpy as np
# 분리된 데이터마다 class 분포가 어떻게 되나
np.unique(y_test, return_counts = True)
```

```python
# 분류 비율을 맞추는 옵션, stratify를 데이터 분리할 때 추가
# 옵션여부, 가중치값, 부여하는 순간 등은 모두 디자이너에 달려있음
X_train, X_test, y_train, y_test = train_test_split(features,
                                                   labels,
                                                   test_size = 0.2,
                                                   stratify = labels,
                                                   random_state =13
                                                   )

np.unique(y_test, return_counts = True)
```

### Decision Tree 함수
- `DecisionTreeClassifier(max_depth = , random_state = )`
    - max_depth
        - 모델을 단순화할 정도로 조정함
        - 조건따라 분화하는 것의 최댓값을 지정

```python
from sklearn.tree import DecisionTreeClassifier
# 학습할 때 일관성을 위해 random_state만 고정
# 모델을 단순화하기 위해 max_depth를 조정
	# 조건따라 분화하는 것의 최대값
iris_tree = DecisionTreeClassifier(max_depth = 2,
                                   random_state = 13)

iris_tree.fit(X_train, y_train)
```

```python
# 조건따라 분화하는 과정 그래프로 보기
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize = (12, 9))
plot_tree(iris_tree)
plt.show()
```

```python
from graphviz import Source
from sklearn.tree import export_graphviz


Source(export_graphviz(iris_tree,
                       feature_names = ['legnth', 'width'],
                       class_names = iris.target_names,
                       rounded = True,
                       filled = True))
```

### accuracy_score
- `from sklearn.metrics import accuracy_score`

```python
# 모델 정확도 확인
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```
